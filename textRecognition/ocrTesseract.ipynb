Text Recognition using Tesseract

```python
import cv2
import pytesseract
from matplotlib import pyplot as plt

pytesseract.pytesseract.tesseract_cmd = "<PATH_TO_TESSERACT>\tesseract.exe"
```

```python
def treatImage(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 0)
    output = cv2.threshold(blur,130,255,cv2.THRESH_BINARY)
    
    return output[1]
```

```python
img = cv2.imread('imgtext1.png')
treatedImg = treatImage(img)
plt.imshow(treatedImg)
```

```python
print(pytesseract.image_to_string(treatedImg))
```
